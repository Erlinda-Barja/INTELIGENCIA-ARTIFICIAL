```python
#Nombre: Barja Coragua Erlinda
#carrera: Ing.de Sistemas
#importacionde librerias
# REPOSITORIO https://github.com/Erlinda-Barja/INTELIGENCIA-ARTIFICIAL
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
from scipy import stats
def generar_datos(n): #se crea una funcion  generara datos para generara los datos aleatorios de peso y estaturas
    datos = [] # se crea array vacio
    for _ in range(n):
        estaturas = random.uniform(0.80, 2.05) # Generar estatura
        # Calcular rango de peso basado en la estatura
        peso_min = 18.5 * (estaturas * estaturas)  # Aproximación simple
        peso_max = 24.9 * (estaturas  *estaturas)
        peso_min = max(40, peso_min - 5)  # No menos de 40 kg
        peso_max = min(120, peso_max + 5)  # No más de 120 kg
        peso = round(random.uniform(peso_min, peso_max), 1)#Se Generar peso dentro del rango
        datos.append((peso, estaturas)) #se añade al vector  satos el peso y la estatura

    return datos #retorna datos

# Generar datos
cantidad = 100
datos = generar_datos(cantidad) #se le pasa la cantidad que se generaran de datos
# Se Imprime la tabla con los datos
print("Peso (kg) | Estatura (m)")
print("-" * 22)
for peso, estaturas in datos:  #se recorre todad los datos
    print(f"{peso:8.1f} | {estaturas:7.2f}")
```

```python
data = pd.DataFrame(datos, columns=['Peso', 'Estatura']) #se crea con la libreria pandas un dataFramer que los organiza  en filas y columnas
x = data['Estatura'] #se guarda las estaturas en este vector x
y = data['Peso']#se guarda los pesos en este vector y
m = np.sum((x - np.mean(x)) * (y - np.mean(y))) / np.sum((x - np.mean(x)) ** 2) #calcula la pendiente  np.mean calcula la media
b = np.mean(y) - m * np.mean(x) #calcula la intercecion

# Crear los valores de y basados en la fórmula de la recta
y_lineal = m * x + b

# Visualización de los datos generados y la recta ajustada
plot.scatter(data['Estatura'], data['Peso'], color='green', label='Datos')# se cre  la grafica de dispercion con los datos de estatura vs peso los puntos lo pinta de color verde
plot.plot(x, y_lineal, color='red', label='Línea ajustada') #se crea la linea ajustada   con lo calculado en y_lineal respecto a la estatura
plot.title('Estatura vs Peso Línea Ajustada')
plot.xlabel('Estatura m')
plot.ylabel('Peso Kg')
plot.legend()
plot.show()
```
